```python
#imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os

#set the path to the CSV file
file_path = "cassan_30_6_1468.csv"
# Microlensing/cassan_30_6_1468.csv

#read the CSV file
data = pd.read_csv(file_path)


x = data['t']
y = data['m']
yerr = data['e']
plt.errorbar(x, y, yerr, fmt = 'o')
plt.xlabel('Time')
plt.ylabel('Magnification')
plt.title('Roman Space Telescope - Simulated Microlensing Light Curve')
plt.show()
```

```python
#zoom in
plt.errorbar(x, y, yerr, fmt = 'o')
plt.xlim(1411, 1416)  # Set x-axis limits
#plt.ylim(0.5, 1.4)  # Set y-axis limits
plt.xlabel('Time')
plt.ylabel('Magnification')
plt.title('Roman Space Telescope - Simulated Microlensing Light Curve (zoomed in)')
plt.show()
```
